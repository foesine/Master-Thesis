```r
evaluation_metrics_factor <- function(predictions, test_set) {
    # Ensure test_set is a data frame
    test_set <- as.data.frame(test_set)
    
    # Ensure both predictions and test_set$income are factors with the same levels
    predictions <- as.factor(predictions)
    reference <- as.factor(test_set$income)
    
    # Ensure levels match between predictions and reference
    levels(predictions) <- levels(reference)
    
    # Confusion matrix for the prediction on original data
    cm <- confusionMatrix(predictions, reference, mode = "everything")

    # Saving evaluation metrics
    accuracy <- cm$overall['Accuracy']
    
    if (length(levels(reference)) == 2) {
        # Binary classification
        f1 <- cm$byClass['F1']
        sens <- cm$byClass['Sensitivity']
        spec <- cm$byClass['Specificity']
    } else {
        # Multi-class classification: calculate metrics for each class and take the mean
        f1 <- mean(cm$byClass[,'F1'], na.rm = TRUE)
        sens <- mean(cm$byClass[,'Sensitivity'], na.rm = TRUE)
        spec <- mean(cm$byClass[,'Specificity'], na.rm = TRUE)
    }

    # Create the dataframe
    metrics_df <- data.frame(
        Accuracy = accuracy, 
        F1 = f1, 
        Sensitivity = sens, 
        Specificity = spec
    )
    
    return(metrics_df)
}
```

```r
discretize_df = function(df, breaks = 5) {
  for (var in colnames(df)) {
    # Check if the variable is not a factor
    if (!is.factor(df[[var]])) {

      # Count the frequency of each unique value
      freq_table <- table(df[[var]])

      # Calculate the proportion of zeros, ensuring NA is handled
      zero_proportion <- ifelse(!is.na(freq_table[as.character(0)]), 
                                freq_table[as.character(0)] / sum(freq_table), 
                                0)

      # Determine the number of breaks based on zero proportion
      if (zero_proportion > 4/5) {
        new_breaks = 1
      } else if (zero_proportion > 1/4) {
        new_breaks = breaks - 2
      } else if (zero_proportion > 1/5) {
        new_breaks = breaks - 1
      } else {
        new_breaks = breaks
      }
      
      # Separate zeros and non-zeros
      zero_portion = (df[[var]] == 0)
      non_zero_values = df[[var]][!zero_portion]

      # Discretize non-zero values
      if (length(non_zero_values) > 0) {
        # Calculate breaks for non-zero values
        range_values = range(non_zero_values, na.rm = TRUE)
        breaks_values = seq(range_values[1], range_values[2], length.out = new_breaks + 1)
        
        # Ensure correct number of labels are created
        labels = sapply(1:(length(breaks_values)-1), function(i) 
                        paste("(", breaks_values[i], "-", breaks_values[i+1], "]", sep=""))

        # Use cut to apply these breaks and labels
        discretized_non_zeros = cut(non_zero_values, breaks = breaks_values, labels = labels, include.lowest = TRUE)
        # Combine zero and discretized non-zeros into the original dataframe
        df[[var]] <- factor(ifelse(zero_portion, "0", as.character(discretized_non_zeros)))
      } else {
        # If all values are zero or the number of breaks is zero or negative
        df[[var]] <- factor("0")
      }
    }
  }
  return(df)
}
```

```r
cpdag_to_dag <- function(cpdag) {
  # Convert bnlearn object to adjacency matrix
  adj_matrix <- amat(cpdag)
  
  # Convert adjacency matrix to igraph object
  ig <- graph_from_adjacency_matrix(adj_matrix, mode = "directed")
  
  # Check if it's already a DAG
  if (igraph::is_dag(ig)) {
    return(cpdag)
  }
  
  # Convert CPDAG to DAG by randomly orienting undirected edges
  directed_arcs <- directed.arcs(cpdag)
  undirected_arcs <- undirected.arcs(cpdag)
  
  while (nrow(undirected_arcs) > 0) {
    arc <- undirected_arcs[1, , drop = FALSE]
    cpdag <- set.arc(cpdag, from = arc[1, 1], to = arc[1, 2])
    undirected_arcs <- undirected.arcs(cpdag)
  }
  return(cpdag)
}
```

```r
train_bn <- function(data, algorithm, score = NULL) {
  if (any(is.na(data))) {
    stop("The data contains missing values.")
  }
  
  data <- droplevels(data)  # Drop unused levels
  
  if (algorithm == "hc") {
    if (is.null(score)) {
      bn <- bnlearn::hc(data)  # Default score (if none provided)
    } else if (score == "aic") {
      bn <- bnlearn::hc(data, score = "aic")
    } else if (score == "bic") {
      bn <- bnlearn::hc(data, score = "bic")
    } else if (score == "k2") {
      bn <- bnlearn::hc(data, score = "k2")
    } else {
      stop("Unsupported score for hc algorithm.")
    }
  } else if (algorithm == "tabu") {
    if (is.null(score)) {
      bn <- bnlearn::tabu(data)  # Default score (if none provided)
    } else if (score == "aic") {
      bn <- bnlearn::tabu(data, score = "aic")
    } else if (score == "bic") {
      bn <- bnlearn::tabu(data, score = "bic")
    } else if (score == "k2") {
      bn <- bnlearn::tabu(data, score = "k2")
    } else {
      stop("Unsupported score for tabu algorithm.")
    }
  } else if (algorithm == "gs") {
    bn <- bnlearn::gs(data)
    bn <- bnlearn::cpdag(bn)  # Retain the graph as a CPDAG
  } else if (algorithm == "iamb") {
    bn <- bnlearn::iamb(data)
    bn <- bnlearn::cpdag(bn)  # Retain the graph as a CPDAG
  } else {
    stop("Unsupported algorithm.")
  }
  
  bn_fitted <- bnlearn::bn.fit(bn, data)
  return(bn_fitted)
}
```

```r
# Define a function to evaluate the model using cross-validation
evaluate_bn <- function(data, bn_fitted, target_var) {
  predictions <- predict(bn_fitted, data = data, node = target_var)
  actual <- data[[target_var]]
  accuracy <- mean(predictions == actual)
  print("any NAs in predictions?")
  print(any(is.na(predictions)))
  return(accuracy)
}
```

```r
inspect_model <- function(bn_fitted) {
    print("Fitted model structure:")
    print(bnlearn::modelstring(bn_fitted))
    
    for (node in names(bn_fitted)) {
        print(paste("CPT for node:", node))
        print(bn_fitted[[node]]$prob)
    }
}
```

```r
bn_pred <- function(data, outer_folds, inner_folds, algorithms = c("hc", "tabu"), scores = c("aic", "bic", "k2")) {
  data <- discretize_df(data)
  grid <- expand.grid(algorithm = algorithms, score = scores)
  outer_results <- list()
  outer_cv_folds = caret::createFolds(data$income, k = outer_folds)

  for (i in seq_along(outer_folds)) {
    outer_test_index = outer_cv_folds[[i]]
    outer_testData = data[outer_test_index,]
    outer_trainData  = data[-outer_test_index,]

    # Ensure factor levels in outer_testData match those in the original data
    for (col in colnames(data)) {
      if (is.factor(data[[col]])) {
        outer_testData[[col]] <- factor(outer_testData[[col]], levels = levels(data[[col]]))
        outer_trainData[[col]] <- factor(outer_trainData[[col]], levels = levels(data[[col]]))
      }
    }

    # Reorder columns to match the original data
    outer_testData <- outer_testData[, colnames(data)]

    # Check for missing values
    if (any(is.na(outer_testData))) {
        stop("outer_testData contains missing values. Handle missing data before prediction.")
    }

    print(paste("Outer Fold", i))
    print(paste("Number of rows in outer_testData:", nrow(outer_testData)))

    inner_results <- list()
    inner_cv_folds = caret::createFolds(outer_trainData$income, k = inner_folds)

    for (j in seq_along(inner_folds)){
        inner_test_index = inner_cv_folds[[j]]
        inner_testData = outer_trainData[inner_test_index,]
        inner_trainData = outer_trainData[-inner_test_index,]

      # Ensure factor levels match
      for (col in colnames(data)) {
        if (is.factor(data[[col]])) {
          inner_testData[[col]] <- factor(inner_testData[[col]], levels = levels(data[[col]]))
          inner_trainData[[col]] <- factor(inner_trainData[[col]], levels = levels(data[[col]]))
        }
      }
      for (col in colnames(inner_testData)) {
        if (is.factor(inner_testData[[col]]) && !all(levels(inner_testData[[col]]) %in% levels(inner_trainData[[col]]))) {
            cat("Missing levels in column:", col, "\n")
        }
      }

      inner_results <- list()

    for (algorithm in algorithms) {
        if (algorithm %in% c("hc", "tabu")) {
            for (score in scores) {
                cat("Fitting with algorithm:", algorithm, "and score:", score, "\n")
                result <- tryCatch({
                    bn_fitted <- train_bn(inner_trainData, algorithm, score)
                    
                    # Inspect the fitted model
                    inspect_model(bn_fitted)
                    
                    predictions <- predict(bn_fitted, newdata = inner_testData)
                    
                    accuracy <- evaluate_bn(inner_testData$income, predictions)
                    data.frame(algorithm = algorithm, score = score, accuracy = accuracy)
                }, error = function(e) {
                    cat("Error fitting model with algorithm:", algorithm, "and score:", score, "\n", e$message, "\n")
                    data.frame(algorithm = algorithm, score = score, accuracy = NA, error_message = e$message)
                })
                
                if (!is.null(result)) {
                    inner_results[[paste(algorithm, score, sep = "_")]] <- result
                }
            }
        } else {
            cat("Fitting with algorithm:", algorithm, "\n")
            result <- tryCatch({
                bn_fitted <- train_bn(inner_trainData, algorithm)
                
                # Inspect the fitted model
                inspect_model(bn_fitted)
                
                predictions <- predict(bn_fitted, newdata = inner_testData)
                
                accuracy <- evaluate_bn(inner_testData$income, predictions)
                data.frame(algorithm = algorithm, score = NA, accuracy = accuracy)
            }, error = function(e) {
                cat("Error fitting model with algorithm:", algorithm, "\n", e$message, "\n")
                data.frame(algorithm = algorithm, score = NA, accuracy = NA, error_message = e$message)
            })
            
            if (!is.null(result)) {
                inner_results[[algorithm]] <- result
            }
        }
    }
    } # End of inner loop

    # Convert the list of results to a data frame
    if (length(inner_results) > 0) {
        inner_results_df <- do.call(rbind, inner_results)
    } else {
        stop("No valid inner results. The list is empty.")
    }

    # Check if inner_results_df is correctly populated
    print("Inner results:")
    print(inner_results_df)

    # Ensure inner_results_df is not empty
    if (nrow(inner_results_df) == 0) {
        stop("No valid inner results. Check if inner_results is correctly populated.")
    }

    # Select the best model
    best_model <- inner_results_df[which.max(inner_results_df$accuracy),]

    # Check if best_model is correctly set
    print("Best model selected:")
    print(best_model)
    
    final_model <- train_bn(outer_trainData, best_model$algorithm, best_model$score)

    # Check outer_testData before prediction
    if (nrow(outer_testData) == 0) {
        stop("outer_testData is empty. Something went wrong with data splitting.")
    }

    predictions <- predict(final_model, newdata = outer_testData)

    eval <- evaluation_metrics_factor(predictions, outer_testData)
    
    outer_results[[i]] <- eval
  } # End of outer loop
  
  eval_avg_outer_folds <- do.call(rbind, outer_results) %>%
                          summarise(across(everything(), mean, na.rm = TRUE))

  return(eval_avg_outer_folds)
} # End of function
```

```r
# Define a function to fit Bayesian networks for different algorithms and scores
fit_bn_full_data <- function(data, algorithms = c("hc", "tabu", "gs", "iamb"), scores = c("aic", "bic", "k2")) {
    results <- list()

    data <- discretize_df(data)

    for (algorithm in algorithms) {
        for (score in scores) {
            # Only apply scores to algorithms that accept them
            if (algorithm %in% c("hc", "tabu")) {
                # Fit the Bayesian network
                cat("Fitting with algorithm:", algorithm, "and score:", score, "\n")
                bn <- tryCatch({
                    train_bn(data, algorithm, score)
                }, error = function(e) {
                    cat("Error fitting model with algorithm:", algorithm, "and score:", score, "\n", e$message, "\n")
                    return(NULL)
                })

                if (!is.null(bn)) {
                    results[[paste(algorithm, score, sep = "_")]] <- bn
                }
            } else {
                # For algorithms that don't use scores
                cat("Fitting with algorithm:", algorithm, "\n")
                bn <- tryCatch({
                    train_bn(data, algorithm)
                }, error = function(e) {
                    cat("Error fitting model with algorithm:", algorithm, "\n", e$message, "\n")
                    return(NULL)
                })

                if (!is.null(bn)) {
                    results[[algorithm]] <- bn
                }
            }
        }
    }

    return(results)
}
```
